```python
import numpy as np
import matplotlib.pyplot as plt
import cv2


def plot_wireframe(projections, image):
	'''
	projections: n_samples X 2 ((height * width) X 2D point coordinate)
	image: on which wireframe has to be plotted
	'''

	# generate 9 points between projections[0] and projections[3]
	points = np.linspace(projections[0], projections[3], 10)
	plt.scatter(points[:, 0], points[:, 1], color='r', s=13, lw=1)

	points2 = np.linspace(projections[1], projections[4], 10)
	plt.scatter(points2[:, 0], points2[:, 1], color='r', s=13, lw=1)

	points3 = np.linspace(projections[2], projections[5], 10)
	plt.scatter(points3[:, 0], points3[:, 1], color='r', s=13, lw=1)

	# generate 7 points between projections[0] and projections[1]
	points4 = np.linspace(projections[0], projections[1], 8)
	plt.scatter(points4[:, 0], points4[:, 1], color='r', s=13, lw=1)

	points5 = np.linspace(projections[3], projections[4], 8)
	plt.scatter(points5[:, 0], points5[:, 1], color='r', s=13, lw=1)

	points6 = np.linspace(projections[1], projections[2], 8)
	plt.scatter(points6[:, 0], points6[:, 1], color='r', s=13, lw=1)

	points7 = np.linspace(projections[4], projections[5], 8)
	plt.scatter(points7[:, 0], points7[:, 1], color='r', s=13, lw=1)




	# draw horizontal lines connecting points and points2
	for i in range(10):
		plt.plot([points[i][0], points2[i][0]], [points[i][1], points2[i][1]], color='r', lw=1)

	# draw horizontal lines connecting points2 and points3
	for i in range(10):
		plt.plot([points2[i][0], points3[i][0]], [points2[i][1], points3[i][1]], color='r', lw=1)
	
	# draw vertical lines connecting points4 and points5
	for i in range(8):
		plt.plot([points4[i][0], points5[i][0]], [points4[i][1], points5[i][1]], color='r', lw=1)
	
	# draw vertical lines connecting points6 and points7
	for i in range(8):
		plt.plot([points6[i][0], points7[i][0]], [points6[i][1], points7[i][1]], color='r', lw=1)


	#draw a scatter plot of all points
	plt.scatter(projections[:, 0], projections[:, 1], color='r', s=13, lw=1)

	plt.imshow(image)

	

	
	




projections = np.array([
	[427, 1726],
	[1549, 1596],
	[2619, 1734],
	[552, 3460],
	[1544, 3003],
	[2494, 3495],

])
image = cv2.imread(
	'/home/praty/Documents/coursework/3-2/CV/assignment-1-spratyey/images/Section-1/calib-object.jpg', cv2.IMREAD_COLOR)
plot_wireframe(projections, image)
```
